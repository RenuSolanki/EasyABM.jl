```julia
using Pkg, Revise
Pkg.activate(joinpath(@__DIR__, ".."))
using EasyABM
```

```julia
model = create_3d_model(size = (5,5,5), threshold = 5)
```

```julia
function initialiser!(model)
    for k in 1:model.size[3]
        for j in 1:model.size[2]
            for i in 1:model.size[1]
                num = rand()
                if num<0.33
                    model.patches[i,j,k].color = :red # stone => red, paper => green, scissor => blue
                elseif num>0.66
                    model.patches[i,j,k].color = :green
                else
                    model.patches[i,j,k].color = :blue
                end
            end
        end
    end
end
init_model!(model, initialiser = initialiser!, props_to_record = Dict("patches" => [:color]))
```

```julia
const sps_dict = Dict(:red => :green, :green => :blue, :blue => :red)
```

```julia
function step_rule!(model)
    for _ in 1:10
        i = rand(1:model.size[1])
        j = rand(1:model.size[2])
        k = rand(1:model.size[3])
        nbr_patches = neighbor_patches_moore((i,j,k), model, 1)
        col = model.patches[i,j,k].color
        rival_col = sps_dict[col]
        count = 0 
        for patch in nbr_patches
            if model.patches[patch...].color == rival_col
                count+=1
            end
        end
        if count > model.parameters.threshold
            model.patches[i,j,k].color = rival_col
        end
    end
end
```

```julia
run_model!(model, steps = 20, step_rule = step_rule!)
```
